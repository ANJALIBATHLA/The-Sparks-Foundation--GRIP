### 1.) Importing data

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
sns.set()
```

```python
#importing dataset from sklearn
raw_data=datasets.load_iris()
data= pd.DataFrame(raw_data.data , columns= raw_data.feature_names)
data.head()
```

```python
#summary of dataset
data.describe()
```

This is a small dataset. There are no missing values.
___

### 2.) Finding optimum number of clusters for k- means clustering

```python
#saving dataset as array
x=data.to_numpy()

from sklearn.cluster import KMeans
wcss=[]
for i in range(1,21):
    kmeans=KMeans(n_clusters=i, init='k-means++',max_iter=300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

#plotting the results
#the elbow method
plt.plot(range(1,21), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

#### Zooming into graph

```python
plt.plot(range(1,11), wcss[:10])
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within cluster sum of squares
plt.show()
```

From this graph, we can see that WCSS value does not decrease significantly after number 3.
* So, we choose number 3.
___

### 3.) Creating K-means classifier

```python
kmeans = KMeans(n_clusters=3, init='k-means++',max_iter=300, n_init=10, random_state=0)
y_clustered = kmeans.fit_predict(x)
```

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_clustered == 0, 0], x[y_clustered == 0, 1], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_clustered == 1, 0], x[y_clustered == 1, 1], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_clustered == 2, 0], x[y_clustered == 2, 1],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

```python
# Visualising the clusters - On last two columns
plt.scatter(x[y_clustered == 0, 2], x[y_clustered == 0, 3], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_clustered == 1, 2], x[y_clustered == 1, 3], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_clustered == 2, 2], x[y_clustered == 2, 3],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:,3], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

```python
# Visualising the clusters - On middle two columns
plt.scatter(x[y_clustered == 0, 1], x[y_clustered == 0, 2], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_clustered == 1, 1], x[y_clustered == 1, 2], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_clustered == 2, 1], x[y_clustered == 2, 2],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:,2], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

```python
# Visualising the clusters - On first and last columns
plt.scatter(x[y_clustered == 0, 0], x[y_clustered == 0, 3], 
            s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_clustered == 1, 0], x[y_clustered == 1, 3], 
            s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_clustered == 2, 0], x[y_clustered == 2, 3],
            s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,3], 
            s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
